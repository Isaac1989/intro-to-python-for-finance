```python
# We will need lots of data for plotting.
# To save time lets create some data quickly.
# This routine creates x's and ys, where the ys are random.

def create_data():
    xs = []
    ys = []
    minN = 2
    maxN = 50
    
    for i in range(10):
        x = i
        # Generate random integers between two numbers
        y = np.random.randint(minN, maxN)
        xs.append(x)
        ys.append(y)
    return xs, ys
# Test routine
x,y = create_data()
print x
print y
```

```python
# We first learn how to plot using functions:
plt.plot(x, y)

```

1. We will generate the time series of two stock price paths

$$S_{t+\Delta t} = S_te^{(\alpha - \sigma^2/2)*\Delta t  + \sigma*sqrt(\Delta t )z_{t+\Delta t}$$

2. Do this for 10 years with dt = 1/12. Do this for two stocks.

3. $\alpha_1 = \alpha_2 = 0.05$ and $\sigma_1 = \sigma_2 = 0.30$, and $S_0 = 100.0$ for both stocks.

4. Store the prices and the  time.

5.Plot the two paths.
6. To simulate ipts + 1 by 2 matrix of  standard normal random variable  use np.random.standard_normal((ipts+1,2))
                        

```python
# Lets create two time series of simulated stock prices
ipts = 120
dt = 1.0/12.0
r = 0.05
sigma = 0.30
y = np.random.standard_normal((ipts+1,2))
S = np.zeros((ipts+1,2))
x = np.zeros(ipts+1)
S[0,0]= 100
S[0,1]= 100
x[0] = 0
for t in range(1,ipts+1):
    S[t,0]= S[t-1,0]*np.exp((r - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*y[t,0])
    S[t,1]= S[t-1,1]*np.exp((r - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*y[t,1])
    x[t]= t*dt

plt.plot(x,S[:,0], label = 'One')
plt.plot(x, S[:,1], label = 'Two')
```

```python
# You have more control as follows

plt.figure(figsize = (7,4)) # Here we divide the whole picture into a 7 by 4 division
plt.grid(True)
plt.plot(S[:,0],'r',label = 'stock A') # 'r' means red line color
plt.plot(S[:,0], '.', color ='b')
plt.plot(S[:,1],'b',label = 'stock B') # 'k' means black line color
plt.plot(S[:,1], '.', color = 'k' )
plt.legend(loc = 0) # loc = 0 means to display the legend on the most suitable place
plt.xlabel ('time')
plt.ylabel ('value')
plt.title ('example')
```

# Controlling the Range  
1.We can configure the ranges of the axes using the set_ylim and set_xlim methods in the axis object, or
axis('tight') for automatically getting "tightly fitted" axes ranges:

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range")
```

## Graphs with separate y axes

```python
#  Here we use two vertical axes
# Lets create two time series of simulated stock prices
def sim2(Start, ipts):
    dt = 1.0/12.0
    r = 0.05
    sigma = 0.30
    y = np.random.standard_normal((ipts+1,2))
    S = np.zeros((ipts+1,2))
    x = np.zeros(ipts+1)
    S[0,0]= Start[0]
    S[0,1]= Start[1]
    x[0] = 0
    for t in range(1,ipts+1):
        S[t,0]= S[t-1,0]*np.exp((r - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*y[t,0])
        S[t,1]= S[t-1,1]*np.exp((r - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*y[t,1])
        x[t]= t*dt
    return x, S
    
x, S = sim2((100,10),120 )    



# Now plot the data     
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(S[:, 0], 'g', lw=1.5, label='A')
ax.set_xlabel("Time")
ax.set_ylabel("A")
ax.legend(loc=4)
#ax.grid()

ax2 = ax.twinx()
ax2.plot(S[:, 1], '--', color ='k', label = 'B')
ax2.set_ylabel("B")
ax2.legend(loc = 0)
```

```python
# Consider the case of two separate subplots. THis gives you more flexibility
plt.figure(figsize=(7, 5))
plt.subplot(211) # This means 2 rows, one column and the first figure ( top one)
plt.plot(S[:, 0], lw=1.5, label='1st')
plt.plot(S[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
plt.title('A Simple Plot')

plt.subplot(212) # THis is the second subplot which will appear in the second row and first column ( ie underneath the first)
plt.plot(S[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(S[:, 1], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
# tag: matplotlib_9
# title: Plot with two sub-plots
```
