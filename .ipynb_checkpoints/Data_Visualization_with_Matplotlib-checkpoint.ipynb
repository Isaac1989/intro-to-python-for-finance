````python
# We need to import a package for data visualization
#%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

# This line of code is used to display pictures in Jupyter Notebook
% matplotlib inline 
figsize = (8, 6)

````

````python
y1 = np.array([5,4,2,1])
y2 = np.array([3,3,1,0])

fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111)
line1, = ax.plot(y1, 'k^-', 
                 markersize = 10, 
                 linewidth = 2,
                 markerfacecolor = 'red',
                 markeredgecolor = 'r',
                label = 'line1')
line2, = ax.plot(y2, 'ms-',
                 markersize = 10, 
                 linewidth = 2,
                 markerfacecolor = 'g',
                 markeredgecolor = 'g',
                label = 'line2')
fig.suptitle('Main title goes here', fontsize = 20, fontweight = 'bold' )

fontdict =  {'fontsize':20, 'fontweight': 'regular'}

ax.set_title('Subtitles/Main titles goes here',fontdict = fontdict)
ax.set_xlabel('x-axis label goes here', fontdict=fontdict)
ax.set_ylabel('y-axis label goes here', fontdict = fontdict)
ax.annotate('Markers', xy = (1,3), xytext = (0.5,1), fontsize = 20,
            arrowprops= {'facecolor':'black', 'shrink':0.05} )

ax.annotate('Markers', xy = (2,2), xytext = (0.5,1),fontsize = 20,
            arrowprops= {'facecolor':'black', 'shrink':0.05} )

ax.annotate('Legend', xy = (3,4.7), xytext = (2,4),fontsize = 20,
            arrowprops= {'facecolor':'black', 'shrink':0.05} )

ax.legend()
plt.show()
````

````python
# We will need lots of data for plotting.
# To save time lets create some data quickly.
# This routine creates x's and ys, where the ys are random.

def create_data():
    xs = []
    ys = []
    minN = 2
    maxN = 50
    
    for i in range(10):
        x = i
        # Generate random integers between two numbers
        y = np.random.randint(minN, maxN)
        xs.append(x)
        ys.append(y)
    return xs, ys
# Test routine
x,y = create_data()
print x
print y
````

## Contrive Exampe in finance

1. We will generate the time series of two stock price paths
$$ S_{t+\Delta t} = S_te^{(\alpha - \sigma^2/2)*\Delta t  + \sigma*\sqrt(\Delta t )}z_{t+\Delta t}$$

2. Do this for 10 years with $dt = \frac{1}{12}$. Do this for two stocks

3. $\alpha_1 = \alpha_2 = 0.05$ and $\sigma_1 = \sigma_2 = 0.30$, and $S_0 = 100.0$ for both stocks.

4. Store the prices and the time.

5. Plot the two paths.

6. To simulate $ipts + 1$ by 2 matrix of  standard normal random variable  use `np.random.standard_normal((ipts+1,2))`

````python
# Lets create two time series of simulated stock prices
ipts = 120
dt = 1.0/12.0
r = 0.05
sigma = 0.30
y = np.random.standard_normal((ipts+1,2))
S = np.zeros((ipts+1,2))
x = np.zeros(ipts+1)
S[0,0]= 100
S[0,1]= 100
x[0] = 0
for t in range(1,ipts+1):
    S[t,0]= S[t-1,0]*np.exp((r - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*y[t,0])
    S[t,1]= S[t-1,1]*np.exp((r - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*y[t,1])
    x[t]= t*dt

plt.figure(figsize = figsize)
plt.plot(x,S[:,0], label = 'One')
plt.plot(x, S[:,1], label = 'Two');
````

# Controlling the axis limits
1.We can configure the ranges of the axes using the set_ylim and set_xlim methods in the axis object, or
axis('tight') for automatically getting "tightly fitted" axes ranges:

````python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range")
````

## Graphs with separate y axes

````python
#  Here we use two vertical axes
# Lets create two time series of simulated stock prices
def sim2(Start, ipts):
    dt = 1.0/12.0
    r = 0.05
    sigma = 0.30
    y = np.random.standard_normal((ipts+1,2))
    S = np.zeros((ipts+1,2))
    x = np.zeros(ipts+1)
    S[0,0]= Start[0]
    S[0,1]= Start[1]
    x[0] = 0
    for t in range(1,ipts+1):
        S[t,0]= S[t-1,0]*np.exp((r - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*y[t,0])
        S[t,1]= S[t-1,1]*np.exp((r - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*y[t,1])
        x[t]= t*dt
    return x, S, y
    
x, S, y = sim2((100,10),120 )    



# Now plot the data     
fig = plt.figure(figsize = figsize)
ax = fig.add_subplot(111)
ax.plot(S[:, 0], 'g', lw=1.5, label='A')
ax.set_xlabel("Time")
ax.set_ylabel("A")
ax.legend(loc=4)
#ax.grid()

ax2 = ax.twinx()
ax2.plot(S[:, 1], '--', color ='k', label = 'B')
ax2.set_ylabel("B")
ax2.legend(loc = 0)
````

````python
# Consider the case of two separate subplots. THis gives you more flexibility
plt.figure(figsize=figsize)
plt.subplot(211) # This means 2 rows, one column and the first figure ( top one)
plt.plot(S[:, 0], lw=1.5, label='1st')
plt.plot(S[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
plt.title('A Simple Plot')

plt.subplot(212) # THis is the second subplot which will appear in the second row and first column ( ie underneath the first)
plt.plot(S[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(S[:, 1], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
# tag: matplotlib_9
# title: Plot with two sub-plots
````

# SEABORN
Seaborn is a high level python library built atop matplotlib. It ships with built-in themes and very reasonable default setting that makes data visualization of less of a hasle. We are going to explore the functiononality of seaborn starting with exploratory examples of what you can do with it. see [__gallery__](https://seaborn.pydata.org/examples/index.html)

### AESTHETICS
Two parts to it
1. Controlling the style of the axes
    1. `axes_style()`
    2. `set_style()`
    
2. Controlling the size of the figure, popularly known as scaling of the plot
    1. `plotting_context()`
    2. `set_context()`

````python
x = np.random.normal(size = (100,6)) + np.arange(6)*0.2
````

## Using matplotlib's default settings to plot boxplots

````python
import seaborn as sns

plt.rcdefaults()
sns.boxplot(data = x);
````

## Seaborn's defaults

````python
sns.set()
sns.boxplot(data = x);
````

## Options for styling the axes

the values that functions axes_style, set_style can take:
1. darkgrid
2. whitegrid
3. white
4. dark
5. ticks

````python
with sns.axes_style('whitegrid'):
    sns.boxplot(data = x, notch = True);
````

# Scaling of plots
In seaborn, the functions `plotting_context` and `set_context` are used for scaling the figure. This is similar to setting the `figsize` property of the canvas/figure object. it takes the following values
1. `talk`, if the graph ismeant to be embed in a talk
2. `poster`, if the graph is meant to embed in a poster
3. `notebook`, if it going to stay in your jupyter notebook or any notebook
4. `paper`, if the graph is going to embed in a paper or pdf


````python
with sns.plotting_context('paper'):     #Try out the other values of plotting_context()
    with sns.axes_style('whitegrid'):
        sns.boxplot(data = data)
````

### A note about customization
The four functions: `axes_style`,`set_style`, `plotting_context`, and `set_context` all have take an argument `rc`
which is a dictionary. The items of this dictionary can be obtained by invoking `sns.axes_style` or `sns.plotting_context`. Further customization can be done using the `sns.set()` function, which take every customization of matplotlib.

````python
sns.axes_style() # or sns.plotting_context()
````

# Color palettes

Seaborn provides us with __color palettes__ just like matplotlib does through it's subpackage `color`, which gives us access to __colormaps__. The computer represent color in a sequence of RGB but humans perceive color via `lightness`, `saturation` and `hue`. These three components are available in seaborn for tweaking various color palettes.

````python
#Current color palettes
current_palette = sns.color_palette()    # a list of tuple of RGBs
sns.palplot(current_palette)   # plotting the color palette
````

````python
# Tweaking lightness and saturation of color palettes
sns.palplot(sns.hls_palette(8, l = 0.3, s = 0.5))
````

## The main families of discrete palettes

1. Qualitative color palettes

    Mainly for representing a data of categorical variable without inherent order
2. Sequential color palettes

    For representing data from less intereting values to more interesting ones or vice versa
    
3. Diverging color palettes

   For representing data values which has both interesting low and high values
   
you can find various color palettes [here](http://matplotlib.org/users/colormaps.html)

````python
# this plot specifically picks colors from the "husl" system

with sns.color_palette('husl', 8):
    plt.scatter(x = np.random.normal(size = 100), y = np.random.normal(size =100))
````

````python
# We can set the color palette 
with sns.axes_style('white'):
    sns.set_palette('husl', 8)
    plt.scatter(x = np.random.normal(size = 100), y = np.random.normal(size =100));
````

````python
# Back to matplotlib's default settings
plt.rcdefaults()
plt.scatter(x = np.random.normal(size = 100), y = np.random.normal(size =100));
````

# Plotting functions in seaborn

### Visualizing the distribution of datasets

````python
# data to be used: alcohol consumption by country
import pandas as pd
drinks=pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
````

### Histogram
Seaborn has the `displot` function plotting the distribution of univariate numerical variable. by defaults it also plots the kernel density function. However, it has parameters to controlling various aspect of the plot. with histograms it is always good to play around with the number of bins as different values may reveal different values in the data

````python
beer_servings = drinks.beer_servings
sns.distplot(beer_servings);
````

````python
# without the kernel dennsity curve (kde)
# the `rug`parameter is set to True to mark the data points with lines
sns.distplot(beer_servings, bins = 20, kde = False, rug= True);
````

### Kernel Density Estimation
This an alternative to the histogram. It is smooth version of the histogram that aims to give the distribution of the variable

````python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (10,8))
sns.kdeplot(beer_servings, ax = ax1);    # without rugplot
sns.kdeplot(beer_servings, ax = ax2, shade = True);  #with rubplot
sns.rugplot(beer_servings, ax= ax2)
````

# Visualizing Categorical variable

````python
# This data set is imdb rating of movies
movies=pd.read_csv('http://bit.ly/imdbratings')
movies.head()
````

## Strip plots
This plot is mainly for plotting quantitative variable segmented into various categories. Take the `star_rating` of the `movie` data. It can be grouped by genre.

````python
sns.stripplot(movies.star_rating, movies.genre);
````

Sometimes the data point may overlap thus obscuring the true pattern. Let look at an example of this phenomenon

````python
tips = sns.load_dataset('tips')
tips.head()
````

````python
#overlaped data points
sns.stripplot('sex', 'tip', data  = tips)
````

````python
# without overlap
sns.stripplot('sex', 'tip', data = tips, jitter = True)
````

````python
# Alternative to avoiding overplotting/overlapse
sns.swarmplot('sex', 'tip', data = tips);
````

````python
# We make a stripplot or swarmplot and then color it another variable using the "hue" argument
sns.swarmplot('day', 'tip', hue = 'sex',data = tips);
````

### Comparing distribution among categories of a numerical variable

The strip and swarm plots are great but the limit our scope of determining the distribution in a particular variable or even compare the distribution among the categories. Thus we use the __`boxplot`__ or **`violin`** plots

````python
sns.set_palette('husl',12)
sns.boxplot('star_rating', 'genre', data = movies);
````

````python
# Another approach is the voilin plots
# Useful for giving insight into the distribution of the various categories
sns.violinplot('day', 'tip', data = tips);
````

````python
# We Can even interlay various plots to gain more insights
sns.set_palette('Paired')
sns.violinplot('day', 'tip', data = tips);
sns.swarmplot('day', 'tip', data = tips, color = 'g');
````

````python
# If the hue parameter has only two levels, we can split the violin plot to obtain the various distribution 

sns.violinplot('day', 'tip', data = tips, hue = 'sex', split = True, inner = 'stick');
````

## Graphing categorical variable segmented by another categorical variable

````python
titanic = sns.load_dataset('titanic')
````

````python
# This give statiscal extimation of the mean count and the corresponding confidence interval
sns.barplot('sex', 'survived', data = titanic, hue = 'class');
````

````python
# Alternative to the barplot but without the bar
sns.pointplot('sex', 'survived', data = titanic, hue = 'class', palette = {'First':'m', 'Second':'g','Third':'b'});
````

````python
# For a regular bar plot 
sns.countplot('survived', data = titanic, hue = 'class');
````
