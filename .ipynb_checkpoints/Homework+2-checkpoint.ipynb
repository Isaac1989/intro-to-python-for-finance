### problem 4

#### Factors

Factors. Write a function called `getfactors()` that takes a single integer as an
argument and returns a list of all its factors, including 1 and itself.

#### Solution

````python
# Solution 1
def getfactors(num):
    """This routine returns the factors of the number `num`
    input
    =====
    num: integer or float
    
    
    return
    ======
    a list of factors of num 
    """
    out = []
    for fac in range(1, num + 1):
        if not (num % fac):
            out.append(fac)
    return out
#getfactors(34)

# Solution  [Alternative 1]
def getfactors_(num):
    """ 
    This routine is an alternative to `getfactors`. it is to demonstrate that a given problem can be solved
    via multiple approach. You just have to choose the optimal one
    """
   
    out, i = [], 1
    
    while i < num + 1:
        if not (num % i):
            out = out + [i]
        i += 1
            
    return out
            
# print getfactors_(16)

# Solution [Alternative 2]
def get_factors(num):
    """ 
    This routine is an alternative to `getfactors`. it is to demonstrate that a given problem can be solved
    via multiple approach. You just have to choose the optimal one
    """
    fac = range(1, num + 1)
    i = 0
    while i < len(fac):
        if num % fac[i]:
            del fac[i]
            i = 0
        else: i += 1
            
    return fac

# get_factors(24)
````

Determine the total number of

2. consonants (counting multiplicity)
3. words (separated by spaces) in the `peom` above. 

Ignore special cases for vowels and
consonants such as "h," "y," "qu," etc.

#### Solution

````python
poem = """ Little bunny Foo Foo
Went hopping through the forest
Scooping up the field mice
And bopping them on the head
Down came the Good Fairy, and she said
"Little bunny Foo Foo
I don't want to see you
Scooping up the field mice
And bopping them on the head.
I'll give you three chances,
And if you don't behave, I will turn you into a goon!" """

# Convert `poem` to lower case
poem = poem.lower()

#Convert string to list a process it to remove unwanted punctuations
poem_list = [x for x in list(poem) if not x in [' ', '\n', '"', ',',"'",'.','!']]

# Count the occurence of each letter
counts = {}
for letter in poem_list:
    counts[letter] = counts.setdefault(letter, 0) + 1
    
vowels = ['i','o','u','a', 'e']

total_vowels = 0
total_consonants = 0 

for letter, count in counts.items():
    if letter in vowels:
        total_vowels += count
    else:
        total_consonants += count
        
print "total vowels: ", total_vowels
print "total consonants: ", total_consonants

# Counting the number of words
words = poem.strip().split(' ')

for word in words:
    if '\n' in word:
        words.extend(word.split('\n'))
        words.remove(word)
    elif "'" in word:
        words.extend(word.split("'"))
        words.remove(word)
        
print 'total words: ', len(words)
````

### Problem 13
Write a function to return Fibonacci Sequence numbers. Fibonacci sequence numbers are as follows: 1, 1, 2, 3, 5, 8, 13, 21, 34, ...From the 3rd number, each nubmer is the summation of previous two numbers. 

#### Solution

````python
def fibonacci(n):
    out = []
    
    def fib(x):
        if x == 1 or x == 2:
            return 1
        else:
            return fib(x-1) + fib(x-2)
        
    for i in range(1, n+1):
        out.append(fib(i))
    return out

# fibonacci(30)
````

````python
#This can be done in  a much more optim way
def my_sort(iterable):
    out = []
    length = len(iterable)
    for i in range(length):
        small_num = min(iterable)
        out.append(small_num)
        iterable.remove(small_num)
    return out

my_sort([50, 23, 4,34])
````

### Problem 15
Suppose you download shares outstanding data from web pages. However, sometimes it is quoted in millions and it ends with "m", such as "100m". Sometimes it is quoted in billions and it ends with "b", such as "100b". Sometimes it is quoted in shares, such as "1000000". Please write a function that can find and substitute "b" and "m" with right numbers. Your function _`expand`_ should return right shares numbers. 

#### Solution

````python
def expand(x):
    """
    This routine expands strings like 100m or 100b to 100000000 and 100000000000
    
    input
    =====
    x: string
    
    return
    ======
    float
    """
    if x.endswith('m'):
        return float(x.replace('m', '0'*6))
    
    elif x.endswith('b'):
        return float(x.replace('b', '0'*9))
    else:
        raise TypeError('The prefix is unknown for proper expansion')

# expand('2b')
````

### Problem 18

Using the previous function, write a new function *`eliminate_max`* that takes in a variable number of arguments and returns a list of its argument with its maximal elements removed.

```python
def eliminate_max(*s):
    """Return a a list containing non-maximal elements of a sequence.

    >>> eliminate_max(1, 3, -1, 3)
    [1, -1]
    """
```

#### Solution

````python
def eliminate_max(*s):
    """Return a a list containing non-maximal elements of a sequence."""
    s = list(s)
    largest = max(s)
    max_rep = s.count(largest)
    for i in range(max_rep):
        eliminate_element(s, max(s))
    return s
eliminate_max(1, 3, -1, 3)
````

### Problem 21

1. Write a function called `is_palindrome` that determines if a string input is palindromic (the same backward as it is forward) 

```python

>>> # Usage
>>> is_palindrome('kayak')
True
>>>is_palindrome('finance')
False

```

2. Adapt the code to take list of strings that return a list of booleans. 

```python
>>>#Usage
>>>is_palindrome('kayak')
True
>>>is_palindrome(['kayak', 'finance'])
False

```

#### Solution

````python

def is_palindrome(string):
    tmp = list(string.lower())
    if tmp == list(reversed(tmp)):
        return True
    else: return False
        
# is_palindrome('finance')
# is_palindrome('kayak')
````

### Problem 23
Given an integer value, Write a function `num2str` that returns a string with the equivalent English text of each
digit. For example, an input of 89 results in `eight-nine` being returned. Extra credit:
Return English text with proper usage, i.e., `eighty-nine.` For this part of the exercise,
restrict values to be between 0 and 1,000

#### Solution

````python


def num2str(x):
    x_str = str(x)
    length = len(x_str)
    digits ={
        0: 'zero',
        1: 'one',
        2: 'two',
        3: 'three',
        4: 'four',
        5: 'five',
        6: 'six',
        7: 'seven',
        8: 'eight',
        9: 'nine',
        10: 'ten',
        11: 'eleven',
        12: 'twelve',
        13: 'thirteen',
        14: 'fourteen',
        15: 'fifteen',
        16: 'sixteen',
        17: 'seventeen',
        18: 'eighteen',
        19: 'nineteen'
    }
    
    prefix100 = {
      1: 'ten',
      2: 'twenty',
      3: 'thirty',
      4: 'forty',
      5: 'fifty',
      6: 'sixty',
      7: 'seventy',
      8:'eighty',
      9: 'ninety'
    }
    
    def two_digits(x):
        x_str = str(x)
        if x_str[1] == '0':
            return prefix100[int(x_str[0])]
        else: 
            return  prefix100[int(x_str[0])] + '-' + digits[int(x_str[1])]
        
    
    if length == 1 or  (x_str.startswith('1') and length <= 2):
        return digits[x]
    
    elif length == 2:
        return two_digits(x)
    
    elif length == 3:
        if x_str[2] == '0' and x_str[1] == '0':
            return digits[int(x_str[0])] + '-hundred '
        
        else:
            if 1<= int(x_str[-2:]) <= 9:
                return digits[int(x_str[0])] + ' hundred and ' + digits[int(x_str[-1])]
            else:
                return digits[int(x_str[0])] + ' hundred and ' + two_digits(int(x_str[-2:]))
    else: 
        return 'One-thousand'
    
    
for i in range(1001):
    print num2str(i)
````
